```python
#  See special TOkens [cls] and [sep]

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "AM LEARNING AGENTIC AI, GEN AI, Deep Learning"

# Use the tokenizer as a callable to get input_ids with special tokens
# This is equivalent to encode_plus for most purposes and often preferred.
encoded_inputs = tokenizer(text, add_special_tokens=True)
input_ids = encoded_inputs["input_ids"]

# Convert input_ids to tokens
tokens = tokenizer.convert_ids_to_tokens(input_ids)
print(tokens)
```

```python
#generate embeddings

from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text =" AI is amaaizing"

inputs = tokenizer(text, return_tensors="pt")

outputs = model(**inputs)

last_hidden_states = outputs.last_hidden_state

print(last_hidden_states.shape)
```
